```python
import pandas as pd
```

```python
df=pd.read_csv('income.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
km.fit(df[['Age','Income($)']])
```

```python
km.labels_
```

```python
df['Clusters']=km.labels_
```

```python
df.head()
```

```python
sns.scatterplot(df,x='Age',y='Income($)',hue='Clusters',palette='Set1',s=100)
```

```python
from sklearn.preprocessing import MinMaxScaler  
scaler=MinMaxScaler()

df_scaled=pd.DataFrame(columns=['Age','Income($)'])

df_scaled[['Age','Income($)']]=scaler.fit_transform(df[['Age','Income($)']])
df_scaled.head()
```

```python
df_scaled['clusters']=km.labels_
```

```python
df_scaled.head()
```

```python
sns.scatterplot(df_scaled,x='Age',y='Income($)',hue='clusters',palette="viridis",s=100)
plt.show()
```
